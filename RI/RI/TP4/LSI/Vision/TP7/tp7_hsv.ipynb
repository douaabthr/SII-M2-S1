```python
import cv2 
import numpy as np
img = cv2.imread('image.png',cv2.IMREAD_COLOR)
img_b=np.zeros(img.shape,img.dtype)
img_g=np.zeros(img.shape,img.dtype)
img_r=np.zeros(img.shape,img.dtype)
#  cv2 BGR ---------
#  une facon de faire 
# img_g[...,0:] 3eme dimension
# img_g[...,0] 3eme dimension c =0
#  img[...,[0,2]] = 0  garder que c = 1 
# img [...,:2] = 0 garder que deux coleur premier 
img_b[:,:,0]=img[:,:,0]
img_g[:,:,1]=img[:,:,1]
img_r[:,:,2]=img[:,:,2]

img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
img_h=np.zeros(img.shape,img.dtype)
img_s=np.zeros(img.shape,img.dtype)
img_v=np.zeros(img.shape,img.dtype)

img_h[:,:,0]=img[:,:,0]
img_s[:,:,1]=img[:,:,1]
img_v[:,:,2]=img[:,:,2]
# imshow lire seulemenet BGR
cv2.imshow('img',img)
cv2.imshow('img hsv ',img_hsv)
cv2.imshow('img b',img_b)
cv2.imshow('img g',img_g)
cv2.imshow('img r',img_r)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
